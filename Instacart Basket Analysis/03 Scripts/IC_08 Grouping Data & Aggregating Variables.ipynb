### 02 Finding the mean for a single column in a dataframe.

```python
# Finding Aggregated mean for number of orders for each department over the entire df.
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']}).map(lambda x: f"{x:0.2f}")
```

### 04 Adding a flag

##### Target marketing criteria: For each user, average price of products < 10 is Low spender; average price of products >= 10 is High spender. Field for calculation of average price (cust_avg_price) will be added. Flag to be added based on comparison to cust_avg_price - "price_spender" (not specified by client).

```python
ords_prods_merge['cust_avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.head()
```

```python
# Adding flag based on criteria compared to cust_avg_price
ords_prods_merge.loc[ords_prods_merge['cust_avg_price'] < 10, 'price_spender'] = "Low spender" 
```

```python
ords_prods_merge.loc[ords_prods_merge['cust_avg_price'] >= 10, 'price_spender'] = "High spender" 
```

```python
# Checking that flag was added properly
ords_prods_merge['price_spender'].value_counts(dropna = False)
```

```python
# Checking the top 40 records for the added flag.
ords_prods_merge[['user_id', 'cust_avg_price', 'price_spender']].head(40)
```

### 05 Adding a second flag

##### Customer frequency criteria -- based on days_since_prior_order for each customer, if the median is: >20, Non-frequent customer; >10 and <=20, Regular customer; <=10, Frequent customer. The column titles were not specified by the customer. I will use cust_med_days for the interim step and cust_frequency for the flag.

```python
# Interim step
ords_prods_merge['cust_med_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Checking Interim step
ords_prods_merge[['user_id', 'cust_med_days']].head()
```

```python
# Adding flag based on criteria compared to cust_med_days
ords_prods_merge.loc[ords_prods_merge['cust_med_days'] > 20, 'cust_frequency'] = "Non-frequent customer" 
```

```python
ords_prods_merge.loc[(ords_prods_merge['cust_med_days'] > 10) & (ords_prods_merge['cust_med_days'] <= 20), 'cust_frequency'] = "Regular customer" 
```

```python
ords_prods_merge.loc[ords_prods_merge['cust_med_days'] <= 10, 'cust_frequency'] = "Frequent customer" 
```

```python
ords_prods_merge.loc[ords_prods_merge['cust_med_days'].isna(), 'cust_frequency'] = "Only one order"
```

#### On check, I found 5 lines with a median of NaN. I found that all 5 lines were from the first order of a customer that had no additional orders in the dataset. I added a condition to assign "Only one order" to those 5 lines to account for the NaN values - they were not errors. The adjusted checks follow.

```python
# Checking cust_frequency flag
ords_prods_merge['cust_frequency'].value_counts(dropna = False)
```

```python
print(ords_prods_merge.loc[ords_prods_merge['cust_frequency'] == 'Only one order'])
```
