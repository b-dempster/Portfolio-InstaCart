### 03 Exclusion of columns with sensitive data

#### Based on the objective of the analysis, it appears that First Name and Surnam columns from the customer dataset are unneccessary and may be left out of the analysis dataset. Excluding this data will protect customer privacy. 
#### Several columns in the customers dataset will be renamed to make them consistent - Gender will be gender, STATE will be state, and Age will be age. For clarity, n_dependants will be additional_family, and fam_status will be marital_status.
#### Also,  busiest_day from the orders_products_merged dataset will be omitted as unneccessary. The flag _merge_ will be excluded to allow a new _merge flag to be generated. 
#### It appears that user_id could be used as a key field to merge the two datasets. The data type in both sets is int64; however, after the merge the data type of this column could be changed to str since this is not quantitiative data. 

```python
# Dropping columns from df_customer to protect customer privacy. The data is not needed for our current analysis.
df_customers = df_customers.drop(columns = ['First Name', 'Surnam'])
```
